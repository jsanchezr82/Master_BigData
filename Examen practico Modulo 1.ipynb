```python
# FunciÃ³n que carga el fichero de Sentimientos en un diccionario Python
def LoadSentimientos (FicSent):
    Fsentimientos = open(FicSent) 
    Dsentimientos = {} 
    for linea in Fsentimientos:
        termino, valor = linea.split("\t")
        Dsentimientos[termino.lower()] = int(valor) 
    return Dsentimientos

# FunciÃ³n que carga el fichero de Tweets en un diccionario Python
def LoadTweets (FicTw):
    Tweets = {}
    with open(FicTw) as Ftweets:
        Tweets = [json.loads(linea) for linea in Ftweets]
    return Tweets

# FunciÃ³n que realiza realiza el ejercicio 1
def Ejercicio (Dsentimientos, Tweets):

    Dtext = {}
    y = 0
    result = {}
    # Cargamos los tweets del fichero en un diccionario
    for Dtweet in Tweets: 
        # Filtramos los tweets sin contenido del diccionario, ya que estos no tienen la clave text
        if "text" in Dtweet: 
            # Creamos una serie con la descomposicion de los textos por palabras y los metemos en un diccionario de series
            Dtext[y] = Series(Dtweet["text"].split()) 
            Numsi = 0
            Numno = 0
            Emocion = 0
            Ltextno = []
            
            for Palabra in Dtext[y][:]:
                # Buscamos cada palabra de la serie en el diccionario de sentimientos formateando las palabras del tweet
                # Se pueden aplicar infinitos formatos para afinar la busqueda, pero lo reducimos para el ejercicio a lower
                # para cubrir el mÃ¡ximo con el mÃ­nimo esfuerzo, pero se podrÃ­an sustituir simbolos y signos de puntuaciÃ³n 
                # por espacios o emogis (por ejemplo ðŸ˜­\ude2d por laugh), etc.
                # En los sentimientos hay algunas palabra compuestas, pero en los tweets es residual, pero tambiÃ©n se pueden
                # evaluar, por ejemplo con un diccionario individual de palabras de sentimientos complejos que busque la cadena
                # en el texto sin explitear ej: good stuff, si good stuff vale 3 y good vale 1 le sumas 2  al sentimiento, 
                # para cuado busques luego por palabra suelta le sumes el 1 restante al encontrar good, en vez de 
                # sumar erroneamente 3 + 1.
                if Palabra.lower() in Dsentimientos:
                    # Contabilizamos la emocion del tweet agregando el resultado en la variable Emocion
                    Emocion += Dsentimientos[Palabra.lower()]
                    # Contamos las palabras coincidentes
                    Numsi += 1
                else:
                    # Ejemplo de sustitucion de Emogis por sentimientos, se podrÃ­a crear un diccionario de Emogis.
                    if Palabra == 'ðŸ˜­' or Palabra == '\ude2d':
                        Emocion += Dsentimientos["laugh"]
                        Numsi += 1
                    else:
                        # Contamos las palabras no coincidentes.
                        Numno += 1
                        # Creamos una lista con las palabras no coincidentes para el Ejercicio 2
                        Ltextno.append(Palabra)

            # Guardamos en un dicionario de series, los textos de los tweets, el nÃºmero de ocurrencias 
            # positivas, el nÃºmero de ocurrencias negativas, el valor del sentimiento, una serie con el tweet descompuesto
            # y la posiciÃ³n de las ocurrencias negativas dentro del tweet descompuesto en un diccionario
            # de series con clave numÃ©rica secuencial.
            result[y] = Series([Dtweet["text"],Numsi,Numno,Emocion,Ltextno])
            
            y += 1

    return result
    
if __name__ == "__main__":
    
    ### MÃ³dulo de Sentimental Analysis
    
    import json
    import numpy as np
    import pandas as pd
    from pandas import *
    
    print("Bienvenido al mÃ³dulo Sentimental Analysis\n")
    
    ## IntroducciÃ³n de los ficheros de sentimientos y tweets y validamos que sean los correctos
    while True:
        print("Introduzca el nombre del archivo de sentimientos a valorar:\n")
        sent = input()
        if sent != "Sentimientos.txt":
                print("Fichero no vÃ¡lido.")
        else:
            FicSentimientos = sent
            break
    
    while True:
        print("Introduzca el nombre del archivo de Tweets a valorar:\n")
        sent = input()
        if sent != "Tweets.txt":
                print("Fichero no vÃ¡lido.")
        else:
            FicTweets = sent
            break
    
    Dsentimientos = {}
    Tweets = {}
    
    # Se carga el fichero de Sentimientos en un diccionario
    
    Dsentimientos = LoadSentimientos(FicSentimientos)
    
    # Se carga el fichero de Tweets en otro diccionario
    
    Tweets = LoadTweets (FicTweets)
    
    print("Â¿Desea ejecutar el ejercicio 1 o el ejercicio 2?\n Teclee 1 o 2: ")
    
    while True:
        Ejer = input()
        if (Ejer.isdigit() and (Ejer == '1' or Ejer == '2')):
            break
        else:
            print("Introduzca un valor vÃ¡lido numÃ©rico entre 1 y 2:")
            
    ## Ejercicio 1
    if (Ejer == '1'):
        
        # Llamamos a la funciÃ³n Ejercicio1 que nos devuelve los textos de los tweets, el nÃºmero de ocurrencias 
        # positivas, el nÃºmero de ocurrencias negativas, el valor global del sentimiento del tweet y una lista de palabras
        # con las palabras sin valor sentimiental del tweet, en un diccionario de series con clave numÃ©rica secuencial.
        Dejercicio1 = Ejercicio (Dsentimientos, Tweets)
        # Extraemos las series de cada clave
        for x in Dejercicio1.values():
            # Pintamos por pantalla el resultado del ejercicio 1.
            print("EL SIGUIENTE TWEET: '" + x[0] + "' TIENE UN SENTIMIENTO ASOCIADO DE: " + str(x[3]) + "\n")
            pass
    
    ## Ejercicio 2
    elif (Ejer == '2'):
        
        # Reutilizamos la funciÃ³n Ejercicio1 que nos devuelve los textos de los tweets, el nÃºmero de ocurrencias 
        # positivas, el nÃºmero de ocurrencias negativas, el valor global del sentimiento del tweet y una lista de palabras
        # con las palabras sin valor sentimiental del tweet, en un diccionario de series con clave numÃ©rica secuencial. 
        # En esta ocasion usaremos las ocurrencias negativas y positivas y la emociÃ³n para calcular el valor medio del 
        # sentimiento del tweet.
        Dejercicio2 = Ejercicio (Dsentimientos, Tweets)
        
        
        for x in Dejercicio2.values():
            #pintamos por pantalla el resultado del ejercicio 2 y calculamos la media al vuelo por cada Tweet.
            if (x[3] > 0): 
                print("Palabras no coincidentes en el Tweet "+ x[0] +":\n")
                for y in x[4]:
                    print("'" + y + "': " + str((x[1]+x[2])/x[3]) + "\n")
                
            
    
    
    
    
```
